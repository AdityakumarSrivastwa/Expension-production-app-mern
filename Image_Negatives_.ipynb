```python
import cv2
```

```python
import numpy as np
```

```python
from google.colab.patches import cv2_imshow
```

**# Load your chosen sample image**

```python
image = cv2.imread('/content/Image1.jpeg')
```

**# Calculate the image negative**

```python
negative_image = 255 - image
```

**# Display the original and negative images**

```python
combined_image = np.hstack((image, negative_image))
```

```python
 cv2_imshow(combined_image)
```
