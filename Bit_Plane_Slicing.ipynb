```python
import cv2
```

```python
import numpy as np
```

```python
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/Image1.jpeg')
```

```python
def bit_plane_slicing(image, bit):
    # Extract the bit-plane
    bit_plane = (image >> bit) & 1
    # Convert the bit-plane to 255 (white) and 0 (black)
    bit_plane = bit_plane * 255
    return bit_plane
```

```python
bit_plane_images = []
```

**# Apply bit-plane slicing for each bit (from 0 to 7) and store the images**

```python
for bit in range(8):
    bit_plane_image = bit_plane_slicing(image, bit)
    bit_plane_images.append(bit_plane_image)
```

```python
combined_images = np.hstack(bit_plane_images)
```

```python
cv2_imshow(combined_images)
```
